```python
import pandas as pd
idata=pd.read_csv('titanic.csv', na_values=['NIL'])
idata.head()
```

```python
import seaborn as sb
sb.pairplot(idata.dropna(),hue="Survived")
```

```python
import matplotlib.pyplot as plt
plt.pie([idata.loc[(idata['Sex'] == 'male') & idata['Survived'] == 0]['Sex'].count(),
         idata.loc[(idata['Sex'] == 'male') & idata['Survived'] == 1]['Sex'].count()]
        ,labels=['Dead Male', 'Survived Male']
       )
plt.axis('equal')
```

```python
plt.pie([idata.loc[(idata['Sex'] == 'female') & idata['Survived'] == 0]['Sex'].count(),
         idata.loc[(idata['Sex'] == 'female') & idata['Survived'] == 1]['Sex'].count()]
        ,labels=['Dead Female', 'Survived Female']
       )
plt.axis('equal')  
```

```python
import matplotlib.pyplot as plt
plt.pie([idata['Sex'].tolist().count('male'),idata['Sex'].tolist().count('female')]
        ,labels=['Male', 'Female']
       )
plt.axis('equal')
```

```python
idata[['Sex','Survived']].groupby(['Sex']).mean().plot.bar()
sb.countplot('Sex',hue='Survived',data=idata,)
plt.show()
```

```python
sb.countplot('Pclass', hue='Survived', data=idata)
plt.show()
```

```python
plt.scatter(idata["Survived"], idata["Age"], alpha=0.1)
```

```python
idata['Fare'].plot(kind='density')
plt.show()
```

```python
import seaborn as sns
grid = sns.FacetGrid(idata, row='Pclass', col='Sex', size=2.2, aspect=1.6)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
```

```python
sb.countplot('Siblings/Spouses Aboard', hue='Survived', data=idata)
plt.show()
```

```python
fare_by_sex = idata.groupby('Sex')['Fare'].mean()
ax = fare_by_sex.plot.bar(title='Fare Average and Sex')
ax.set_ylabel('Average Fare')
```
